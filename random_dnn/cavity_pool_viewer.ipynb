# Live scatter of the cavity resolvent diagonal

Each dot is one member of the population of resolvent diagonal entries
G_ii maintained by the cavity population dynamics of
`RMT.cavity_svd_resolvent` (blue = g1, green = g2 family), at one
singular value `s` and fixed pool size -- the exact production numba
kernel, rendered client-side via ipycanvas at up to 60 fps. The viewer
starts paused: press `continue`. All knobs (alpha, sigma_W, s, pop,
eta, phi, dtype, axis scale/limits, steps/frame, trails) apply live
between sweeps without restarting the run; see the `LiveCanvas`
docstring for the full reference. If the canvas is blank in VS Code,
add jsdelivr.com under the "Jupyter: Widget Script Sources" setting.

```python
import sys
from pathlib import Path

# the module is a sibling of this notebook (random_dnn/)
if str(Path.cwd()) not in sys.path:
    sys.path.insert(0, str(Path.cwd()))
from cavity_pool_viewer import LiveCanvas

live = LiveCanvas(
    alpha=1.5,
    sigma_W=1.0,
    s=2.0,        # singular value; try s past the edge (~3+) for collapse
    pop=128,      # one frame per sweep: pop<=256 -> 55-60 fps, 512 -> ~15
    # steps_per_frame=32,  # big pools (up to 8192): sub-sweep frames keep fps up
)
live.show()  # starts paused; re-running this cell stops the previous run
```
